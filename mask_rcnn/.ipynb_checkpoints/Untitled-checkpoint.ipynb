```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def maskres(mask):
    mask = mask.astype(np.uint8)  #convert to an unsigned byte
    mask*=255
    G = np.full((mask.shape[0],mask.shape[1]), 0, dtype=np.uint8)
    B = np.full((mask.shape[0],mask.shape[1]), 0, dtype=np.uint8)
    maskReshaped = np.dstack((B,G,mask))
    return maskReshaped
    
```

```python
img = cv2.imread('bird.jpg',1)
mask =np.load('mask_rcnn_mask.npy')
mask_reshaped = np.reshape(mask,(mask.shape[0],mask.shape[1]))
#mask = mask.astype(int)
alpha = 0.5
color = [0,0,255]
```

```python
def add_mask(img,mask_reshaped):
    for c in range(3):
    img[:, :, c] = np.where(mask_reshaped == True,
                              img[:, :, c] *
                            (1 - alpha) + alpha * color[c] * 255,
                            img[:, :, c])
    return img
```

```python
for c in range(3):
    img[:, :, c] = np.where(mask_reshaped == True,
                            img[:, :, c],
                              img[:, :, c] *
                            (1 - alpha) + alpha * color[c] * 255)
```

```python
dst = cv2.addWeighted(img,0.7,mask_reshaped,0.3,0)
```

```python
cv2.imshow('Indices',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
